### Scarp data from wikipedia and clean it over

```python
import requests
import pandas as pd 
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
table = pd.read_html(url)[0]

df = table[table['Borough'] != 'Not assigned']
df = df.groupby(['Postcode', 'Borough']).agg(', '.join).reset_index()
df.rename(columns={ df.columns[0]: "PostalCode" }, inplace = True)
df
```

```python
df.shape
```

```python
geo_df = pd.read_csv('Geospatial_Coordinates.csv')
geo_df.head()
```

```python
df_merge = pd.merge(df, geo_df, on='PostalCode')
df_merge
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(df_merge['Borough'].unique()),
        df_merge.shape[0]
    )
)
```

```python
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
import folium
```

```python
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_merge['Latitude'], df_merge['Longitude'], 
                                           df_merge['Borough'], df_merge['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```

```python
Scarborough_data = df_merge[df_merge['Borough'] == 'Scarborough'].reset_index(drop=True)
Scarborough_data.head()
```

```python
address = 'Scarborough, Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Scarborough are {}, {}.'.format(latitude, longitude))
```

```python
map_scarborough = folium.Map(location=[latitude, longitude], zoom_start=11)

for lat, lng, label in zip(Scarborough_data['Latitude'], Scarborough_data['Longitude'], 
                           Scarborough_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_scarborough)  
    
map_scarborough
```
