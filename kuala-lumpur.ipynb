```python
address = 'Bukit Bintang, kuala-lumpur'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
limit = 100
radius = 1000
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET,VERSION ,latitude, longitude, radius, limit)
url
```

```python
results = requests.get(url).json()
results
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
venues
```

```python
venues_normalized = json_normalize(venues)
venues_normalized
```

```python
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
venues_normalized = venues_normalized.loc[:, filtered_columns]
venues_normalized
```

```python
venues_normalized['venue.categories'] = venues_normalized.apply(get_category_type, axis=1)
venues_normalized
```

```python
venues_normalized.columns = [col.split(".")[-1] for col in venues_normalized.columns] 
venues_normalized.tail()
```

```python
map_newyork = folium.Map(location=[latitude, longitude], zoom_start=15)

# add markers to map
for lat, lng, venues in zip(venues_normalized['lat'], venues_normalized['lng'], 
                            venues_normalized['name']):
    
    label = venues
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_newyork)  
    
map_newyork
```

```python
print("There are {} different categories".format(len(venues_normalized['categories'].unique())))
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
venues_onehot = pd.get_dummies(venues_normalized['categories'])
venues_onehot.head()
```

```python
k = 15

kmeans = KM(n_clusters=k, random_state=55).fit(venues_onehot)

kmeans.labels_[51:56]
```

```python
venues_normalized['clusters'] =  kmeans.labels_
```

```python
venues_normalized = venues_normalized.astype({'clusters':float})
venues_normalized.head()
```

```python
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=15)

x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(venues_normalized['lat'], venues_normalized['lng'],
                                  venues_normalized['name'], venues_normalized['clusters']):
    label = folium.Popup(poi + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
