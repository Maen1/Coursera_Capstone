```python
address = 'Bukit Bintang, kuala-lumpur'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
limit = 100
radius = 1000
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET,VERSION ,latitude, longitude, radius, limit)
url
```

```python
results = requests.get(url).json()
results
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
venues
```

```python
venues_normalized = json_normalize(venues)
venues_normalized
```

```python
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
venues_normalized = venues_normalized.loc[:, filtered_columns]
venues_normalized
```

```python
venues_normalized['venue.categories'] = venues_normalized.apply(get_category_type, axis=1)
venues_normalized
```

```python
venues_normalized.columns = [col.split(".")[-1] for col in venues_normalized.columns] 
venues_normalized.tail()
```
